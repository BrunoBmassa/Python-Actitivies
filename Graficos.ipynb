```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import gdown
import numpy as np
import textwrap as tw
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

url='https://drive.google.com/file/d/1dit9NB9hI4wKbU_zvGOZNRSTDPTK-dzx/view?usp=drive_link'

# Extração do arquivo pela URL
file_id = url.split('/')[-2]

# Download do arquivo
output_file = 'ecommerce_estatistica.csv'  # You can change the output filename
gdown.download(id=file_id, output=output_file, quiet=False)

# Leitura do arquvio CSV usando pandas
df = pd.read_csv(output_file)

#Grafico 1
#Histograma

#plt.figure(figsize=(6,4))
plt.hist(df['Preço'], bins=100, color = 'red', alpha= 0.8)
plt.title('Histograma Preço')
plt.xlabel('Preço')
plt.ylabel('Frequencia')
plt.grid(True)
plt.show()

#Grafico 2
#Dispersão

plt.subplot(2, 2, 2)
plt.scatter(df['Preço'], df['Preço'], color = 'green', alpha= 0.5, s=20)
plt.title('Dispersão Preço e Preço')
plt.xlabel('Preço')
plt.ylabel('Preço')

#Grafico 3
#Mapa de calor

corr = df[['Preço', 'Qtd_Vendidos_Cod']].corr()
plt.subplot(2,2,3)
#plt.figure(figsize=(10,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação Preço e Qtd_Vendidos_Cod')

#Grafico 4
#Barras

plt.figure(figsize=(35,10))
plt.subplot(2,2,4)
df['Gênero'].value_counts().plot(kind='bar', color='purple')
plt.title('Quantidade de Produtos por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.grid(True)

#Grafico 5
#Pizza

x = df['Temporada'].value_counts().index
y = df['Temporada'].value_counts().values
plt.figure(figsize=(10,6))
plt.pie(y, labels=x, autopct='%.1f%%', startangle=180,)
#counts.plot(kind='pie', labels = None, figsize=(11,8), legend=True)
plt.title('Quantidade de vendidos')
plt.show()

#Grafico 6
#Dispersão

plt.hexbin(df['Nota'], df['N_Avaliações'], gridsize=40, cmap='Blues')
plt.colorbar(label='')
plt.xlabel('Nota')
plt.ylabel('Número Avaliações')
plt.title('Dispersão de Notas e Número de avaliações')
plt.show()

#Grafico 7
#Regressão

sns.regplot(x='Preço', y='Qtd_Vendidos_Cod', data=df, color='yellow', scatter_kws={'alpha':0.5, 'color': 'orange'})
plt.title('Regressão Preço e Qtd_Vendidos_Cod')
plt.xlabel('Preço')
plt.ylabel('Qtd_Vendidos_Cod')
plt.show()

plt.tight_layout()
```
